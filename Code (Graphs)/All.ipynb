```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
import pandas as pd
import matplotlib.ticker
malware = pd.read_csv('malstat.csv')
alltime = pd.read_csv("Alltime.csv")
alltime_products = pd.read_csv("allprod.csv")
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) { //This is a code block that disables scroll
    return false;
}
```

```python
f = plt.figure()
f.set_figwidth(11)
f.set_figheight(7)
ax = plt.axes()
ax.set_facecolor("#221C35")
plt.xlim([2013, 2021])
plt.grid(axis='y', linestyle='-')
plt.title('Malware per OS')
matplotlib.ticker.StrMethodFormatter("{x:.6f}")
plt.plot(malware.Year , malware.Windows, color = "#0078d4")
plt.plot(malware.Year , malware.Android, color = "#78C257")
plt.plot(malware.Year , malware.Macos, color = "#bf00ff")
plt.legend(['Windows' , 'Android', 'MacOS'])
plt.xlabel('year')
plt.ylabel('Malware')
```

```python
tot = pd.read_csv('total.csv')
```

```python
f = plt.figure()
f.set_figwidth(11)
f.set_figheight(7)
ax = plt.axes()
ax.set_facecolor("Black")
plt.grid(axis='y', linestyle='-')
plt.xlim([2017.5, 2022.5])
plt.ylim([0, 1400000000])
plt.bar(tot.Year , tot.Total, color = "Purple")
plt.xlabel('year')
plt.ylabel('Malware (Up to 1.4 Billion)')
plt.legend(['Total Malware'])
plt.title('Total Malware/Year')
ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter("{x:.0f}"))
```

```python
f = plt.figure()
f.set_figwidth(11)
f.set_figheight(7)
ax = plt.axes()
ax.set_facecolor("#221C35")
ax.set_axisbelow(True)
plt.grid(axis='y', linestyle='-')
mw = statistics.mean(malware.Windows)
mm = statistics.mean(malware.Macos)
ma = statistics.mean(malware.Android)
plt.bar("Windows", mw, color = "#0078d4")
plt.bar("MacOS", mm, color = "#bf00ff")
plt.bar("Android", ma, color = "#78C257")
plt.title("Average Malware")
plt.ylabel('Malware (Millions)')
ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter("{x:.0f}"))
```
